## 1. Einleitung

Der Begriff Künstliche Intelligenz (KI, AI) beschreibt Systeme oder Maschinen, die Aufgaben ausführen können, die traditionell menschliche Intelligenz erfordern - etwa das Erkennen von Sprache oder Bildern, das Treffen von Entscheidungen oder das Lösen von Problemen. KI ist dabei ein übergeordneter Begriff, unter dem verschiedene Teilbereiche wie Robotik, Natural Language Processing (NLP) oder auch Maschine Learning zusammengefasst werden können. 

![Übersicht zu Künstlicher Intelligenz](https://www.researchgate.net/publication/354348678/figure/fig1/AS:11431281187572961@1694283946559/Comprehensive-overview-of-artificial-intelligence-AI-and-its-subfields-from.tif) 

Referenz: https://www.researchgate.net/publication/354348678_The_Impact_of_Artificial_Intelligence_and_Deep_Learning_in_Eye_Diseases_A_Review

Machine Learning als zentrales Teilgebiet der künstlichen Intelligenz gewinnt zunehmend an Bedeutung, ob in der Industrie bei Vorhersagen von Fehlerwahrscheinlichkeiten für Maschinenteile, als Entscheidungshilfe für Unternehmen oder als Tool, um unsere Produktivität zu steigern und Entwicklungszeit zu verringern. Mit zunehmender Geschwindigkeit der Entwicklungen im Bereich Künstliche Intelligenz ist es sinnvoll, nicht nur die mathematischen Grundlagen zu verstehen, sondern auch zu wissen, welche Methode wann sinnvoll einsetzbar ist, welche Tools und Libraries uns bei der Umsetzung unterstützen können und wie wir datengetriebene Modelle sauber und effizient in die bestehende Infrastruktur eines Kunden integrieren. 

"Machine Learning" beschreibt im Kern Systeme, die mithilfe verschiedener Algorithmen aus Daten lernen und sich verbessern, ohne dabei explizit dafür programmiert zu werden. Statt auf auf feststehenden Regeln zu basieren, erkennen Machine Learning Modelle Muster und Korrelationen in unübersichtlich großen Datenmengen und nutzen diese, um Vorhersagen zu und Entscheidungen zu treffen. Diese Modelle werden umso genauer, je mehr Daten sie zur Verfügung haben. Ein klassisches Beispiel dafür ist ein Spamfilter, der automatisch E-Mails in "Spam" und "Nicht-Spam" klassifizieren soll. Statt jede Regel per Hand zu definieren und zu programmieren, wird ein Modell mit tausenden markierten Beispielen (Label: Spam/Nicht-Spam) trainiert und lernt so, welche Worte oder andere Merkmale typisch für Spam-Mails sind.

Machine Learning als Sammelbegriff für verschiedene algorithmische Ansätze lässt sich grob in drei Hauptkategorien unterteilen: **überwachtes Lernen**, bei dem Modelle aus gekennzeichneten Daten lernen (z. B. Klassifikation oder Regression), **unüberwachtes Lernen**, das Strukturen in unmarkierten Daten erkennt (z. B. Clustering), sowie **Reinforcement Learning**, bei dem ein Agent durch Belohnungssysteme in einer Umgebung lernt, optimale Entscheidungen zu treffen. Jedes dieser Verfahren hat unterschiedliche Anwendungsbereiche und Stärken – von der Kundensegmentierung bis hin zur automatisierten Bilderkennung oder dem Training intelligenter Chatbots.

#### **[scikit-learn](https://scikit-learn.org/stable/user_guide.html)**

skikit-learn ist die wohl bekannteste und mächtigste Open-Source Bibliothek für Machine Learning. Sie enthält alle Funktionalitäten, um einen kompletten ML-Workflow zu erstellen und stellt viele robuste Algorithmen zum maschinellen Lernen, unter anderem für Clustering, Regression, Klassifikation und Dimensionsreduktion, zur Verfügung. Die Performance eines Modells hängt maßgeblich mit der Qualität der zugrunde liegenden Trainingsdaten zusammen, daher ist das Vorbereiten der Daten ein essentieller Schritt in jeder Analyse.

scikit-learn verfolgt einen konsistenten API-Standard, der es ermöglicht, nahezu alle Modelle auf dieselbe Weise aufzubauen, zu trainieren und zu evaluieren. Dabei werden zentrale Schritte wie Feature-Skalierung, Modell-Pipelines, Cross-Validation, Hyperparameter-Tuning und Modellbewertung unterstützt – alles mit geringem Codeaufwand. Dank der nahtlosen Integration mit Bibliotheken wie pandas, NumPy und matplotlib/Seaborn eignet sich scikit-learn besonders gut für strukturierte Daten, wie sie im Unternehmensumfeld typischerweise auftreten: Kundenstammdaten, Transaktionshistorien, Sensordaten oder Marktanalysen.

Die Bibliothek eignet sich daher hervorragend für schnelle Prototypen und Projekte, in denen Transparenz und Nachvollziehbarkeit eine wichtige Rolle spielen – zum Beispiel in Bereichen wie Risikobewertung, Kundenklassifikation oder Vertriebsprognosen, wo die Entscheidungen eines Modells begründbar sein müssen. Da viele der eingesetzten Algorithmen (wie Entscheidungsbäume, lineare Modelle oder k-Means) mathematisch überschaubar bleiben, lassen sich ihre Ergebnisse gut erklären und visualisieren – ein wichtiger Vorteil in der Kommunikation mit Fachbereichen oder Kunden. Auch regulatorische Anforderungen, wie sie in der Finanz- oder Gesundheitsbranche auftreten, lassen sich mit scikit-learn einfacher erfüllen als mit komplexeren Blackbox-Modellen. Für tiefergehende Modellierung oder Deep-Learning-Anwendungen kann scikit-learn problemlos mit anderen Tools wie TensorFlow oder PyTorch kombiniert werden.

````python
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn import datasets
````

````python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
print(iris_df.head())
print( "Features: ", iris.feature_names)
print("Labels: ", iris.target_names)
print("Dimensionalität: ", iris.data.shape)
sb.scatterplot(data=iris_df, x=iris_df.columns[0], y=iris_df.columns[1], hue='species')
````

scikit-learn hat einige Datensätze integriert, die für Übungszwecke verwendet werden können. Mit dem **Iris-Datenset** werden die Größen von Kron- und Kelchblättern dargestellt sowie die Namen ihrer Spezies hinterlegt, was ihn besonders für Klassifikationsaufgaben interessant macht. Datensatzobjekte sind so aufgebaut, dass in der Eigenschaft .data die Daten in einem Array hinterlegt sind, das die Dimensionalität Anzahl_Samples x Anzahl_Features hat. Dieses lässt sich problemlos in ein pandas Dataframe überführen. Die für überwachtes Lernen relevanten Label, also Klassenbezeichnungen, werden in .target bzw. in .target_names gespeichert.

##### **Funktion 1: Aufsplitten der Daten in Trainings- und Testdaten**

test_size und random_state bestimmen die prozentuale Aufteilung in Trainings- und Testdaten sowie einen Randomisierungsfaktor (zur Reproduzierbarkeit der Ergebnisse)

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_df[iris.feature_names], iris_df['species'], test_size=0.2, random_state=42) 
````

##### **Funktion 2: Feature Scaling (Standardisierung)**

Viele Modelle erwarten eine Standardisierung, da sonst Ergebnisse verzerrt und Ergebnisse nicht sinnvoll verwendet werden können. Jeder Wert wird dann berechnet durch: $ a = (b-c)/d $ wobei:

a ... skalierter Featurewert \
b ... unskalierter Featurewert \
c ... Mittelwert aller Featurewerte \
d ... Standardabweichung der Featurewerte

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
print("Trainingsdaten original: \n")
print(X_train.head())
print("\nTrainingsdaten normalisiert: \n")
print(X_train_scaled[0:5])
````

##### **Funktion 3: One-Hot-Encoding***

Machine-Learning-Algorithmen erwarten numerische Eingabedaten und können mit kategorischen Variablen nicht direkt umgehen. One-Hot-Encoding ist eine gängige Methode, um solche Variablen umzuwandeln: Dabei wird für jede Ausprägung eine eigene Spalte erstellt, die den Wert 1 erhält, wenn die Kategorie zutrifft – und 0, wenn nicht. Diese Methode ist eher ungeeignet, wenn es viele mögliche Ausprägungen der kategorischen Variable gibt. \
Zu Demonstrationszwecken nehmen wir an, 'species' sei kein Label, sondern auch ein Feature.

````python
# Variante 1
iris_encoded = pd.get_dummies(iris_df['species'], dtype=int)
print("Dataframe vorher: \n")
print(iris_df.head())
print("\nDateframe nachher: \n")
print(pd.concat([iris_df[iris.feature_names], iris_encoded], axis=1).head())

# Variante 2
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
species_encoded = encoder.fit_transform(iris_df[['species']])
print("\n One-Hot-Encoding mit scikit-learn\n")
print(species_encoded[0:5])
````

##### **Übersicht Funktion 4: Modelltraining**

Dieser Abschnitt dient nur der Übersicht, wo gängige Modelle gefunden werden können. Erklärungen zu den Modellen erfolgen in einem späteren Abschnitt

````python
# KMeans-Clustering
from sklearn.cluster import KMeans
model_kmeans = KMeans(n_clusters=3, random_state=42)
model_kmeans.fit(X_train)

# Principal Component Analysis
from sklearn.decomposition import PCA
model_pca = PCA(n_components=2)
pca_result = model_pca.fit_transform(X_train)

# Random Forest Klassifikation
from sklearn.ensemble import RandomForestClassifier
model_random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
model_random_forest.fit(X_train, y_train)

# Support Vector Machine
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)

# Deep Learning am Beispiel MLP (Multi-Layer Perceptron) Klassifikator
from sklearn.neural_network import MLPClassifier
model_mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=100, random_state=42)
model_mlp.fit(X_train, y_train)
````

##### **Übersicht Funktion 5: Modellevaluierung und Performance**

````python
# Clustering (ohne Ground-Truth)
# Silhouette Score und Davies Bouldin Index

from sklearn.metrics import silhouette_score, davies_bouldin_score
labels = model_kmeans.labels_
print("Clusteringmetriken:")
print("\nSilhouette Score (je näher an 1 desto besser):", silhouette_score(X_train, labels))
print("Davies-Bouldin Index (je näher an 0 desto besser):", davies_bouldin_score(X_train, labels))

# Klassifikation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
# Beispiel: Auswertung des Random Forest
y_pred = model_random_forest.predict(X_test)

print("\nKlassifikationsmetriken:")
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("F1-Score:", f1_score(y_test, y_pred, average='macro'))
print("\nKlassifikationsreport:\n", classification_report(y_test, y_pred, target_names=iris.target_names))

# Confusion Matrix
print("Konfusionsmatrix:")
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot()
````

#### **[TensorFlow](https://www.tensorflow.org/guide)**

TensorFlow ist ein Open-Source Paket, das von Google entwickelt wurde und vorrangig für Deep Learning Anwendungen eingesetzt werden kann. Es ermöglichst den Aufbau und das Training komplexer neuronaler Netze - von einfachen Feedforward-Modellen bis hin zu anspruchsvollen, mehrschichtigen Architekturen, die in Bereichen von Bild- und Spracherkennung oder in der Textverarbeitung wie dem Natural Language Processing (NLP) zum Einsatz kommen. Durch den integrativen Ansatz lassen sich Modelle mobil oder in Cloud-Umgebungen deployen und anwenden. Dazu kommt eine breite Unterstützung für verschiedene Programmiersprachen wie beispielsweise Go, C++ oder JavaScript.

Ein zentrales Konzept in TensorFlow – insbesondere in der ursprünglichen Version vor dem intensiven Einsatz von Keras – ist das Arbeiten mit sogenannten Dataflow Graphs. Ein Dataflow-Graph besteht aus Knoten (Operations) und Kanten (Tensoren), die den Fluss von Daten (Tensors) durch das Modell definieren. Jeder Knoten repräsentiert eine mathematische Operation, während die Kanten die Daten (multidimensionale Arrays) transportieren.

Dieses Modell ermöglicht es, komplexe Rechenoperationen effizient und übersichtlich zu organisieren und auf verschiedenen Geräten (z. B. CPU, GPU, TPU) auszuführen. Auch wenn mit der Einführung von Keras der direkte Umgang mit Graphen weniger sichtbar geworden ist, bildet dieses Prinzip weiterhin das Rückgrat der TensorFlow-Architektur. Für produktionsnahe Anwendungen, Deployment und Optimierungen (z. B. über tf.function) sind Graphen auch heute noch entscheidend, da sie bessere Performance und Portabilität bieten.

**Typische Funktionalitäten**

Als Beispiel wird wieder das Iris-Datenset verwendet. Die Preprocessing-Schritte sind hier gleich zu denen in scikit-learn.

````python
#!pip install tensorflow
````

````python
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Iris-Daten laden
iris = load_iris()
X = iris.data
y = iris.target

# Features skalieren
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Trainings-/Testdaten splitten
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
````

**Funktion 1: Aufbau eines Modells**

Die mathematischen Hintergründe werden in einem späteren Abschnitt behandelt.

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Einfaches sequentielles Modell
model = Sequential([ # Layer werden nacheinander geschichtet
    Dense(16, activation='relu', input_shape=(X_train.shape[1],)), # Input-Layer
    Dense(8, activation='relu'), # 1 Hidden Layer mit ReLU-Aktivierungsfunktion
    Dense(3, activation='softmax')  # 3 Klassen im Iris-Datensatz, Output-Anzahl muss immer der Klassenanzahl entsprechen
])

# Kompilieren (Optimizer, Loss, Metrik)
model.compile(optimizer='adam', # häufiger Optimizer
              loss='sparse_categorical_crossentropy', # Standard für One-Hot-Encoding
              metrics=['accuracy']) # Bewertung der Modellvorhersage
````

**Funktion 2: TensorBoard**

TensorBoard ist ein Tool innerhalb von TensorFlow, das Trainingsmetriken visualisiert.

````python
# Training
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir="./logs", histogram_freq=1)
````

**Funktion 3: Modelltraining und Evaluierung**

````python
# Standard-Training
train_standard = model.fit(
                    X_train,
                    y_train,
                    epochs=30,
                    validation_data=(X_test, y_test),
                    callbacks=[tensorboard_callback]
)

# Training mit Early Stopping, also Trainingsabbruch nach [patience] Schritten ohne Modellverbesserung

from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='val_loss', patience=3)

print("Early Stopping Training: \n")
train_early = model.fit(
                X_train, 
                y_train, 
                epochs=50, 
                batch_size=8,
                validation_split=0.2, 
                callbacks=[early_stop])
````

##### **Funktion 4: Visuelle Darstellung des TensorBoard mit Trainingsmetriken**

##### **Funktion 5: Schnelle Evaluierung mit Vorhersage**

````python
predictions = model.predict(X_test)
predicted_classes = np.argmax(predictions, axis=1)

print("Vorhergesagte Klassen:", predicted_classes[:10])
model.evaluate(X_test, y_test)

````

*Dataset* ist eine Klasse, die die Daten und ihre zugehörigen Label speichert. \

````python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Transformation: Umwandlung in Tensor 
# und Normalisierung (Zahlen: globaler Mittelwert und Standardabweichung für MNIST)
# Diese Transformationen sind Standard für diesen Datensatz und dienen der Umwandlung
# numerische Daten
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# Training- und Testdaten laden
train_data = datasets.MNIST(root='data', train=True, download=True, transform=transform)
test_data = datasets.MNIST(root='data', train=False, download=True, transform=transform)

# DataLoader erzeugen
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=1000, shuffle=False)
````

##### **Funktion 2: Modelldefinition**

````python
import torch.nn as nn
import torch.nn.functional as F

class MNIST_Model(nn.Module): # Basisklasse aller NN
    def __init__(self):
        super(MNIST_Model, self).__init__() # drei vollvernetzte Layer
        self.fc1 = nn.Linear(28*28, 128) 
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10) # 10 mögliche Klassen (0-9)

    def forward(self, x):
        x = x.view(-1, 28*28) # Bild wird zu Vector
        x = F.relu(self.fc1(x)) # Anwendung der ReLU Aktivierungsfunktion
        x = F.relu(self.fc2(x))
        return self.fc3(x) # Rückgabe ohne Softmax-Umwandlung in Wahrscheinlichkeiten (weitere Schritte wören notwendig)
````

##### **Funktion 3: Trainingsvorbereitung**

````python
model = MNIST_Model()
# NN brauchen zur besseren Performance immer Loss-Funktionen
# Cross Entropy eignet sich für Klassifikationen gut
loss_function = nn.CrossEntropyLoss()
# Adam ist ein Standard Optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
````

##### **Funktion 4: Training**

````python
# im Gegensatz zu TensorFlow werden die Epochs hier direkt geschrieben
for epoch in range(10):
    # Versetzen des Modells in Trainingszustand
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        optimizer.zero_grad()
        # Aufruf einer Instanz der Modellklasse wie eine Funktion
        # ruft unterliegend automatisch die forward() Funktion auf
        output = model(images)
        loss = loss_function(output, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {running_loss:.4f}")
````

##### **Funktion 5: Modellevaluierung**

````python
# Modell in Evaluierungsmodus versetzen
model.eval()
correct = 0
total = 0

# beschleunigt Berechnungen, für Evaluierung werden keine Gradienten gebraucht
with torch.no_grad():
    for images, labels in test_loader:
        output = model(images)
        # Identifikation der vorhergesagten Klassen
        # entspricht dem höchsten Wert des Modelloutputs
        predictions = torch.argmax(output, dim=1)
        correct += (predictions == labels).sum().item()
        total += labels.size(0)

print(f"Test Accuracy: {correct / total:.2f}")
````

## 3. Unüberwachte Lernalgorithmen

Unüberwachtes Lernen ist ein Teilbereich des maschinellen Lernens, bei dem Algorithmen aus nicht gelabelten Daten Muster und Strukturen erkennen sollen. Im Gegensatz zum überwachten Lernen gibt es keine Zielvariable (z.B.: Klasse, Pflanzenart, Autotyp), anhand derer das Modell trainiert wird, sondern das Modell versucht, inhärente Zusammenhänge, Gruppierungen oder Regelmäßigkeiten in den Eingabedaten zu erkennen. 

Typische Fragestellungen lauten:
- Welche Kund:innen verhalten sich ähnlich?
- Welche Produkte werden häufig gemeinsam gekauft?
- Gibt es redundante Dimensionen in einem Datensatz?
- Welche Anomalien treten in medizinischen Daten auf?

Unüberwachte Lernalgorithmen spielen eine zentrale Rolle in der **explorativen Datenanalyse**, im **Feature Engineering**, im **Clustering** bzw. der **Segmentierung** oder in der **Regelidentifikation** und liefern oft den ersten Schritt für datengetriebene Strategien in Unternehmen.

### 3.1 K-Means Clustering

K-Means ist ein Clusteringverfahren, bei dem Datensätze in **K Gruppen/Cluster** unterteilt werden - basierend auf der euklidischen Distanz zum nächsten Cluster. Dabei minimiert der Algorithmus die Summe der Punktdistanzen zum jeweiligen Clusterzentrum (Zentroid). Somit erfolgt eine automatische Einteilung von Datenpunkten in Gruppen.

**Typische Anwendungsbereiche:**
- Kundensegmentierung
- Mustererkennung in Transaktionsdaten
- Datenvorverarbeitung

**Funktionsweise:**

![Prinzip](https://stanford.edu/~cpiech/cs221/img/kmeansViz.png)

(a) Datensatz\
(b) Initialisiere zufällig k Zentroids\
(c) Berechne für jeden Punkt die euklidische Distanz zu allen Zentroids und ordne jeden Punkt dem Zentroid mit der niedrigsten euklidischen Distanz zu\
(d) Berechne einen neuen Zentroid als Mittelwert der zugehörigen Punkte\
(e-f) Wiederhole Schritte 3-5, bis es keine Veränderungen mehr gibtfrom sklearn.datasets import load_iris

#### Beispiel anhang des Iris-Datensatzes

````python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

# Iris-Daten laden
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# KMeans-Clustering mit k=3 (wir kennen die echten Klassen, tun aber so, als wären sie unbekannt)
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
X['cluster'] = kmeans.fit_predict(X)

# Zentroiden extrahieren
centroids = kmeans.cluster_centers_
````

````python
# Für die Darstellung nehmen wir zwei Features
sb.set(style="whitegrid")
plt.figure(figsize=(8,6))
sb.scatterplot(x=iris.data[:, 0], y=iris.data[:, 1], hue=X['cluster'], palette='Set1')
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='black', marker='x', label='Zentroiden')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("K-Means Clustering auf Iris-Daten")
plt.legend()
plt.show()
````

#### Qualitätsmetriken

Die sogenannte **Inertia** misst die Summe der quadrierten Abstände jedes Datenpunktes zu seinem zugeordneten Zentroid. Je größer die Inertia, desto ineffizienter sind die Clusterzentren gewählt und die Anzahl der Zentroiden sollte angepasst werden. Als Faustregel kann man sich merken, dass ein gutes K-Means-Modell eine niedrige Inertia und eine niedrige Anzahl Cluster hat. Mit jedem zusätzlichen Zentroid sinkt zwar *immer* die Inertia, aber die Gefahr des **Overfittings** steigt damit auch, denn die niedrigste Inertia tritt dann auf, wenn jeder Datenpunkt sein eigener Zentroid ist. Auch ist die Inertia nur *innerhalb **eines** Datensatzes* vergleichbar. Zur Auswahl der Clusteranzahl kann die sogenannte **Elbow-Methode** verwendet werden, bei der die Inertia für verschiedene Anzahlen von Zentroiden (k) ermittelt und gegeneinander aufgetragen wird. Die optimale Anzahl liegt meist dort, wo die Inertia weniger stark zu sinken beginnt - dem "Ellbogen" des Graphen. In diesem Beispiel liegt dies bei 3.

````python
print("Inertia (Sum of Squared Distances to Cluster Centers):", kmeans.inertia_)
````

````python
inertias = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    km.fit(iris.data)
    inertias.append(km.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel("Anzahl Cluster (k)")
plt.ylabel("Inertia")
plt.title("Elbow-Methode")
plt.show()
````

### 3.2 Hierarchisches Clustern

Hierbei handelt es sich um einen Algorithmus, der Daten nicht auf einmal, sondern *schrittweise gruppiert*. Dabei entsteht eine baumartige Struktur - Dendrogramm genannt -, das die Beziehungen zwischen Datenpunkten zeigt. So werden ähnliche Datenpunkte sukzessive in Cluster zusammengefasst, ohne vorher die Anzahl der Cluster festlegen zu müssen.

Es gibt zwei Ansätze:
- Agglomerativ (Bottom-Up): Jeder Datenpunkt ist zu Beginn ein eigenes Cluster und diese werden nach und nach zusammengeführt.
- Divisiv (Top-Down): Alle Punkte beginnen in einem Cluster, das schrittweise aufgestalten wird.

Das divisive Clustern wird seltener verwendet. Beide Algorithmen sind greedy, was bedeutet, dass sie bei der Wahl der Zusammenführung oder Teilung von Clustern immer lokale Optima nutzen, wodurch sie nicht immer die beste Lösung erzeugen. Zudem sollte bedacht werden, dass hierarchische Clusteralgorithmen hohe Rechenleistungen beanspruchen.

**Typische Anwendungsbereiche:**
- Kategorisierung von Patientenkohorten in homogenere Gruppen in der medizinischen Forschung
- Gruppierung von handgeschriebenen Zeichen anhand ihrer Form
- Explorative Datenanalyse

**Funktionsweise**
In jedem Schritt wird eine *Distanzmatrix* berechnet, die den Abstand jedes Clusters zu jedem anderen enthält. Dieser Abstand wird mithilfe einer vorher ausgewählten Verknüpfungsmethode (*Linkage*) berechnet. Diese sind die gängigsten Methoden:
- *Single Linkage*: Abstandsberechnung auf Grundlage der Punkte in den Clustern, die am nächsten zusammen liegen (bildet oft schmale, lange Cluster)
- *Complete Linkage*: Abstandsberechnung aufgrund von Punkten, die am weitesten auseinander liegen (bildet runde, kompakte Cluster)
- *Average Linkage*: Abstandsberechnung anhand des durchschnittlichen Abstandes aller Punktepaare zwischen zwei Clustern (wirkt oft am Natürlichsten)
- *Ward*: Zusammenführen von Clustern anhand der minimalen Summe der quadrierten Abstände der Punkte vom Clusterzentrum (Minimierung der Varianz)

#### Für metrische Daten eignet sich die Methode nach **Ward** oft am Besten.

````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import seaborn as sb
import pandas as pd

#Ungleichmäßige Clusterstruktur erzeugen
X, y_true = make_blobs(
    n_samples=[50, 100, 20],        # Ungleich große Cluster
    centers=[[-5, 0], [0, 0], [5, 5]], # Position der Clusterzentren 
    cluster_std=[1.0, 1.5, 0.5],    # Unterschiedliche Streuungen
    random_state=3
)
df = pd.DataFrame(X, columns=["x", "y"])

# Linkage-Methoden zum Vergleich
methods = ["single", "complete", "average", "ward"]

plt.figure(figsize=(14, 8))
for i, method in enumerate(methods):
    plt.subplot(2, 2, i + 1)
    
    Z = linkage(df, method=method)
    cluster_labels = fcluster(Z, t=3, criterion='maxclust')
    df["cluster"] = cluster_labels

    sb.scatterplot(data=df, x="x", y="y", hue="cluster", palette="Set2", s=40)
    plt.title(f"{method.capitalize()} Linkage")
    plt.xlabel("x")
    plt.ylabel("y")

plt.suptitle("Vergleich von Linkage-Methoden (unbalancierte Cluster)", fontsize=16)
plt.tight_layout()
plt.show()
````

### 3.3 Principal Component Analyse

Die PCA oder Hauptkomponentenanalyse ist ein Verfahren zur **Dimensionsreduktion**. Sie transformiert die ursprünglichen Merkmale eines Datensatzes in neue, unkorrelierte Variablen, die sogenannten Hauptkomponenten, ohne dabei zu viel Information zu verlieren. Somit können Redundanzen reduziert und die Daten vereinfacht werden, allerdings sind diese Komponenten nicht intuitiv interpretierbar und haben keine echte Bedeutung, sondern sind nur ein mathematisches Hilfsmittel für folgende Analsen. Der Algorithmus versucht, möglichst viel Information in den Hauptkomponenten unterzubringen, beginnend mit der ersten. 

**Typische Anwendungsbereiche**
- Visualisierung hochdimensionaler Daten
- Vorverarbeitung von Daten für Klassifikation oder Clustering
- Reduktion von Rauschen oder Korrelationen
- Erkennung dominanter Strukturen in Daten

#### *Merke!*
Warum sollten Variablen im maschinellen Lernen unkorreliert sein?
- Redundanzen (korrelierte Variablen tragen ähnliche Informationen)
- Je weniger Variablen, desto stabiler und einfacher zu interpretieren sind Modelle
- Voraussetzung für einige Algorithmen (Naive Bayes, PCA, LDA)
- Je weniger gut differenzierbare Features vorhanden, desto schneller und effizienter trainieren Modelle

**Funktionsweise:**
1. Standardisierung der Daten
2. Berechnung der Kovarianzmatrix zur Ermittlung von Korrelationen
3. Berechnung von Eigenvektoren und Eigenwerten zur Bestimmung der Hauptkomponenten
4. Umformung der Daten entlang der Achsen der Hauptkomponenten (PC)

#### Es gibt immer genauso viele PC wie ursprüngliche Features! 

Aufgrund der Informationskondensierung können in Anschlussberechnungen PC mit niedrigem Informationsgehalt (niedrige Varianz) weggelassen werden (**Dimensionsreduktion**; Faustregel: 95% der Varianz/Information sollte abgedeckt sein). Dabei ist zu beachten: Nachfolgende Algorithmen müssen mit abstrakten Features arbeiten können, da die PC keine direkt interpretierbaren Features mehr sind.


#### Beispiel anhand des Iris-Datensatzes

Varianz = Information

````python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

# Daten laden und skalieren
iris = load_iris()
X = iris.data
X_scaled = StandardScaler().fit_transform(X)

# PCA mit 4 Hauptkomponenten (ein Datensatz hat immer so viele PC wie Variablen)
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_scaled)
````

````python
import pandas as pd
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)],
    index=iris.feature_names
)
print("Kovarianzmatrix der einzelnen Features mit der Hauptkomponente im Interval (-1, 1). Je näher die Zahl an 1 (positive Korrelation) oder -1 (negative Korrelation), desto größer ist der Einfluss des Features auf die Hauptkomponente.")
print(loadings)
````

````python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd

# Loadings (Gewichtungen der Features auf die PCs)
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)],
    index=iris.feature_names
)

# Heatmap anzeigen
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sb.heatmap(loadings, annot=True, cmap="coolwarm", center=0)
plt.title("PCA-Loadings – Einfluss der Features auf Hauptkomponenten")
plt.xlabel("Hauptkomponente")
plt.ylabel("Ursprüngliche Features")
plt.subplot(1, 2, 2)
plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
         pca.explained_variance_ratio_*100, marker='o')
plt.title("Informationsgehalt pro Komponente (Scree-Plot)")
plt.xlabel("Hauptkomponente")
plt.ylabel("Erklärte Varianz/Informationsgehalt (%)")
plt.grid()
plt.tight_layout()
plt.show()
````

### 3.4 A-Priori-Algorithmus

Bei A-Priori handelt es sich um einen Assoziationsalgorithmus, der regelmäßige Muster und Zusammenhänge in Transaktionsdaten finden soll. Er wird also recht spezifisch für Warenkorbanalysen angewendet, sodass häufig gemeinsam gekaufte Produkte identifiziert und schließlich datengetriebene Verkaufsstrategien entwickelt werden können. Andere Anwendungsfälle sind die Erkennung von Zusammenhängen zwischen Symptomen und Krankheiten in der Medizin, Nutzungsverhaltensanalyse oder Betrugserkennung.

**Definitionen**
- Itemset ... eine Menge an Elementen, die aus mindestens 2 Elementen besteht
- Support ... Anteil der Transaktionen, in denen ein Itemset vorkommt
- Confidence ... bedingte Wahrscheinlichkeit eines Itemsets
- Lift ... Verhältnis von beobachteter zu erwarteter Häufigkeit von Item 1 wenn Item 2

Beispiel:\
Itemset: {Brot, Margerine} \
Support = 70% (Brot und Margerine werden in 70% der Fälle gemeinsam gekauft) \
Confidence (Brot -> Margerine) = 80% (*wenn* Brot, *dann* Margerine zu 80% Wahrscheinlichkeit)

Ein entscheidender Nachteil dieses Algorithmus ist seine Ineffizienz, da die Anzahl der möglichen Itemsets exponentiell mit der Anzahl von Items steigt.

**Algorithmusablauf**
1. Join: Bildung von Itemsets der Größe K (K ist der Wiederholungsschritt)
2. Prune: Entfernung aller Itemsets, die eine gesetzten Support-Schwelle nicht erreichen und somit als zu selten angesehen werden

### Manuelles Beispiel

````python
import pandas as pd

transactions = [
    ['Milch', 'Brot', 'Butter'],
    ['Bier', 'Chips'],
    ['Milch', 'Brot'],
    ['Milch', 'Butter'],
    ['Brot', 'Butter'],
    ['Milch', 'Brot', 'Butter'],
    ['Milch', 'Chips'],
]

# One-Hot-Encoding manuell erzeugen
all_items = sorted(set(item for t in transactions for item in t))
df = pd.DataFrame([{item: (item in t) for item in all_items} for t in transactions])
df = df.astype(int)
df
````

````python
item_support = df.mean()
print("Einzel-Support:\n", item_support)
````

````python
from itertools import combinations

min_support = 0.3  # 30 %

frequent_pairs = []
for item1, item2 in combinations(all_items, 2):
    support = ((df[item1] & df[item2]).sum()) / len(df)
    if support >= min_support:
        frequent_pairs.append((item1, item2, support))

pair_df = pd.DataFrame(frequent_pairs, columns=['Item1', 'Item2', 'Support'])
pair_df
````

````python
rules = []

for item1, item2, support_xy in frequent_pairs:
    support_x = item_support[item1]
    support_y = item_support[item2]

    confidence_x_to_y = support_xy / support_x
    confidence_y_to_x = support_xy / support_y
    lift = support_xy / (support_x * support_y)

    rules.append({
        'Antecedent': item1,
        'Consequent': item2,
        'Support': support_xy,
        'Confidence (X→Y)': confidence_x_to_y,
        'Confidence (Y→X)': confidence_y_to_x,
        'Lift': lift
    })

rules_df = pd.DataFrame(rules)
rules_df.sort_values('Lift', ascending=False)
````
